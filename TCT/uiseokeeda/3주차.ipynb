```python
# input 값 받기 
N = int(input())
k = int(input())
k_lists = []
for i in range(k):
    k_list =[int(x) for x in input().split()]
    k_lists.append(k_list)
l = int(input())
l_lists = []
for i in range(l):
    x,c =  input().split()
    l_lists.append([int(x),c])

# 보드 생성(벽까지 생성한다)
def make_board(N):
    # 리스트의 초기 선언은 *으로 하면 복사이기때문에 리스트 함축으로 해줄 것 
    board = [[0 for j in range(N+2)] for i in range(N+2)]
    for i in range(N+2):
        board[0][i] = 3
        board[N+1][i] = 3
    for i in range(N+2):
        board[i][0] = 3
        board[i][N+1] = 3
    return board
    
# 사과 위치만 2로 변경시켜 주는 것     
def make_apple(board,apple_lists):
    for i in apple_lists:
        board[i[0]][i[1]] = 2

# 뱀의 이동
def move(bam,next,apple):
    bam.append(next)
    if(apple):
        return bam
    else:
        # 꼬리 삭제 
        del bam[0]

# 뱀의 죽음
def live(board,bam,next):
    for i in bam:
        if(i==next):
            return False
    if(board[next[0]][next[1]]==3):
        return False    
    return True

# 뱀의 방향전환(현재 방향과 turn방향 받음)
def turn(direction,turn):
    if(turn == 'D'):
        #북
        if(direction[0]==-1 and direction[1]==0):
            return ([0,1])
        #동
        elif(direction[0]==0 and direction[1]==1):
            return ([1,0])
        #남
        elif(direction[0]==1 and direction[1]==0):
            return ([0,-1])
        #서
        else:
            return([-1,0])
    else:
        #북
        if(direction[0]==-1 and direction[1]==0):
            return [0,-1]
        #동
        elif(direction[0]==0 and direction[1]==1):
            return ([-1,0])
        #남
        elif(direction[0]==1 and direction[1]==0):
            return ([0,1])
        #서
        else:
            return([1,0])


board = make_board(N)
make_apple(board,k_lists)
is_live = True
bam = [[1,1]]
direction = [0,1]
second = 0
index = 0

while(is_live):
    if(index<len(l_lists)):
        if(second == l_lists[index][0]):
            direction = turn(direction,l_lists[index][1])
            index+=1

    next_position = [bam[len(bam)-1][0]+direction[0],bam[len(bam)-1][1]+direction[1]]
    print(next_position)
    print(bam)
    if(live(board,bam,next_position)==False):
        is_live = False

    # 사과 먹었을 때
    if(board[next_position[0]][next_position[1]]==2):
        apple = True
    # 사과 못먹었을 때
    else:
        apple = False
    move(bam,next_position,apple)
    
    second+=1

print(second)
```

### 내가 푼 방식

```python
# 기둥과 보 설치
# https://programmers.co.kr/learn/courses/30/lessons/60061
from time import time
# 기둥, 보 삭제함수
def crush(frame,what,answer):
    # 기둥 삭제 조건
    if(what==0):
        can = True
        # 아래에 기둥이 있을 때
        if([frame[0],frame[1]-1,0] in answer):
            can = False
        # 왼 기둥 삭제 시에 보만 존재할 때
        elif([frame[0],frame[1],1] in answer):
            # 다른쪽에 기둥 존재x 
            if(([frame[0]+1,frame[1],0] in answer)==False):
                can = False
        # 오른 기둥 삭제 시에 보만 존재할 때
        elif([frame[0]-1,frame[1],1] in answer):
            # 다른쪽에 기둥 존재x
            if(([frame[0]-1,frame[1],0]in answer) == False):
                can = False
    # 보 삭제 조건
    else:
        can = True
        # 오른쪽 보가 기둥에 없을때
        if ((([frame[0]+1,frame[1],0] in answer)==False) and (([frame[0]+2,frame[1],0] in answer)==False)):
            can = False
        # 왼쪽 보가 기둥에 없을때
        elif ((([frame[0]-1,frame[1],0] in answer)==False) and (([frame[0]-2,frame[1],0] in answer)==False)):
            can = False
        # 보 왼쪽 위에 기둥이 있을 때 
        elif ([frame[0],frame[1],0] in answer):
            # 아래 기둥이 없을 때
            if(([frame[0],frame[1]-1,0] in answer) == False):
                can = False
        # 보 오른쪽 위에 기둥이 있을 때
        elif ([frame[0]+1,frame[1],0]in answer):
            if(([frame[0]+1,frame[1]-1,0] in answer) == False):
                can = False
    
    if(can == True):
        answer.remove([frame[0],frame[1],what])

# 기둥, 보 설치함수(좌표와 어떤것)
def build(frame,what,answer):
    # 기둥 설치 조건
    if(what == 0):
        can = False
        # 바닥일경우 
        if(frame[1]==0):
            can = True
        # 보의 한쪽 끝 부분 위
        elif ([frame[0]-1,frame[1],1] in answer):
            can = True
        # 다른 기둥 위    
        elif ([frame[0],frame[1]-1,0] in answer):
            can = True
    # 보 설치 조건     
    else:
        can = False
        # 왼쪽 끝부분이 기둥 위
        if([frame[0],frame[1]-1,0] in answer):
            can = True
        # 오른쪽 끝부분이 기둥 위
        elif([frame[0]+1,frame[1]-1,0] in answer):
            can = True
        # 양쪽 끝부분이 다른 보와 동시에 연결 
        elif([frame[0]-1,frame[1],1] in answer and [frame[0]+1,frame[1],1] in answer):
            can = True 

    if(can==True):
        answer.append([frame[0],frame[1],what])

def solution(n, build_frame):
    answer = []
    for i in build_frame:
        if(i[3]==1):
            build([i[0],i[1]],i[2],answer)
        else:
            crush([i[0],i[1]],i[2],answer)
    answer.sort(key=lambda x: (x[0], x[1],x[2]))
    return answer

build_frame = [[1,0,0,1],[1,1,1,1],[2,1,0,1],[2,2,1,1],[5,0,0,1],[5,1,0,1],[4,2,1,1],[3,2,1,1]]
n = 5
start_time = time()
print(solution(n,build_frame))
end_time = time()

dur = end_time - start_time
print(dur)
```

### 답지

```python
def possible(answer):
    for x,y,stuff, in answer:
        if stuff==0:
            if y==0 or [x-1,y,1] in answer or [x,y,1] in answer or [x,y-1,0] in answer:
                continue
            return False
        elif stuff == 1:
            if [x,y-1,0] in answer or [x+1,y-1,0] in answer or ([x-1,y,1]) in answer and [x+1,y,1] in answer):
                continue
            return False
        return True
    
def solution(n,build_frame):
    answer = []
    for frame in build_frame:
        x,y,stuff,operate = frame
        if operate == 0:
            answer.remove([x,y,stuff])
            if not possible(answer):
                answer.append([x,y,stuff])
        if operate == 1:
            answer.append([x,y,stuff])
            if not possible(answer):
                answer.remove([x,y,stuff])
    return sorted(answer)
```

```python
# 치킨 배달
from itertools import combinations

N,M = input().split()
N = int(N)
M = int(M)

city = []
for i in range(N):
    a,b,c,d,e = input().split()
    city.append([int(a),int(b),int(c),int(d),int(e)])

houses = []
chickens = []
for i in range(N):
    for j in range(N):
        if(city[i][j]==1):
            houses.append([i+1,j+1])
        elif(city[i][j]==2):
            chickens.append([i+1,j+1])

# 가능한 경우 만들어주기 
chickens = list(combinations(chickens,M))

# 거리 계산 함수 
def distance(chicken,houses):
    result = []
    for i in houses:
        dist = []
        for j in chicken:
            dist.append(abs(i[0]-j[0])+abs(i[1]-j[1]))
        result.append(min(dist))
    return sum(result) 

dist_list = []
for i in chickens:
    dist_list.append(distance(i,houses))

print(min(dist_list))
```

```python
import itertools
def solution(n, weak, dist):
    length = len(weak)
    # 1. weak 배열의 길이를 2배로 늘리기
    for i in range(length):
        weak.append(weak[i]+n)
    # 2. 투입할 친구의 최솟값을 찾기 위해 최댓값보다 1 큰 값으로 초기화
    answer = len(dist)+1
    # 3. 0부터 length-1까지 start 이동하면서 찾기
    for start in range(length):
        for friends in list(itertools.permutations(dist,len(dist))):
            count = 1 # 투입할 친구 수
            # 현재 친구가 점검 가능한 위치 구하기
            position = weak[start]+friends[count-1]
            # 시작점 부터 모든 취약지점 확인
            for i in range(start, start+length):
                if position < weak[i] :
                    count += 1
                    if count > len(dist):
                        break
                    position = weak[i]+friends[count-1]

            answer = min(count,answer)

    if answer > len(dist):
        return -1
    return answer
```

```python
from collections import deque

n,m,k,x = map(int, input().split())
graph = [[] for _ in range(n+1)]

for _ in range(m):
    a,b = map(int,input().split())
    graph[a].append(b)

distance = [-1]*(n+1)
distance[x] = 0

q = deque([x])
while q:
    now = q.popleft()
    for next_node in graph[now]:
        if (distance[next_node] == -1):
            distance[next_node] = distance[now]+1
            q.append(next_node)

check = False
for i in range(1,n+1):
    if(distance[i] == k):
        print(i)
        check = True
        
if check == False:
    print(-1)
        
```
